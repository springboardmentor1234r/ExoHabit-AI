Key Feature Identification

Habitability depends on both planetary and host star properties.

Planetary features such as radius, mass, orbital period, and equilibrium temperature influence surface conditions.

Stellar features such as temperature, luminosity, and spectral type determine the energy received by the planet.

These features are therefore selected as the primary inputs for habitability analysis.

```python
data_dictionary = {
    "pl_name": "Name of the exoplanet",
    "hostname": "Name of the host star",
    "pl_rade": "Planet radius (Earth radii)",
    "pl_masse": "Planet mass (Earth masses)",
    "pl_orbper": "Orbital period (days)",
    "pl_eqt": "Planet equilibrium temperature (K)",
    "pl_orbsmax": "Semi-major axis (AU)",
    "st_teff": "Host star effective temperature (K)",
    "st_lum": "Host star luminosity",
    "st_spectype": "Host star spectral type",
    "st_met": "Host star metallicity"
}

pd.DataFrame(data_dictionary.items(), columns=["Column Name", "Description"])
```

Initial Observations

The dataset contains a wide range of planetary and stellar parameters.

Several important features such as planet mass and temperature contain missing values.

Most discovered exoplanets are gas giants, indicating a class imbalance in habitable candidates.

Stellar temperature and luminosity vary significantly across host stars.

These observations highlight the importance of data cleaning and feature engineering, which will be performed in Week 2.
